### Choose number of cluster

```python
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
```

```python
_X = pd.DataFrame(X)
_X.dropna(inplace= True)
```

```python
pca = PCA(n_components=2)
pca.fit(_X)
x_3d = pca.transform(_X)
```

```python
# Compute the WCSS for different values of K
wcss = []
for i in range(1, 15):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=0)
    kmeans.fit(x_3d)
    wcss.append(kmeans.inertia_)

# Plot the curve of WCSS against the number of clusters K
plt.plot(range(1, 15), wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

Quan sát biểu đồ, áp dụng phương pháp WCSS, ta chọn n_cluster = 4

```python
kmeans = KMeans(n_clusters = 4, init = "k-means++", random_state = 42)
y_kmeans = kmeans.fit_predict(x_3d)
```

```python
plt.scatter(x_3d[y_kmeans == 0, 0], x_3d[y_kmeans == 0, 1], s = 60, c = 'red', label = 'Cluster1')
plt.scatter(x_3d[y_kmeans == 1, 0], x_3d[y_kmeans == 1, 1], s = 60, c = 'blue', label = 'Cluster2')
plt.scatter(x_3d[y_kmeans == 2, 0], x_3d[y_kmeans == 2, 1], s = 60, c = 'green', label = 'Cluster3')
plt.scatter(x_3d[y_kmeans == 3, 0], x_3d[y_kmeans == 3, 1], s = 60, c = 'violet', label = 'Cluster4')

plt.legend()

plt.show()
```

```python
def zscore_scaler(column):
  mean = column.mean()
  std  = column.std()

  return column.apply(lambda x: (x-mean)/std)
```
